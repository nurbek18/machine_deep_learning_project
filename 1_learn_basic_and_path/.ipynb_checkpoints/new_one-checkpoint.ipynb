```python
#matplotlib 的引入和使用
x = [1,2,3,4,5]
y = [4,3,5,6,2]



from matplotlib import pyplot as plt
%matplotlib inline
#plot 是函数图
fig1 = plt.figure(figsize = (7,7))
plt.plot(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# scatter 是散点图
fig1 = plt.figure(figsize = (7,7))
plt.scatter(x,y)
plt.title('y vs x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
