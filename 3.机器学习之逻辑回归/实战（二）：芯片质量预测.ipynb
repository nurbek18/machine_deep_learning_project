#### 1.基于chip_test.csv数据 ， 建立逻辑回归模型（二阶边界），评估模型表现；

```python
#load the data  ; visualize data ; generate new data;
# establish model and train it ; predict 
# accuraacy
#decision boundary
#define f(x)



```

```python
#load the data
import pandas as pd
import numpy as np
data = pd.read_csv("chip_test.csv")
data.head()
```

```python
#add label mask # 创建一个布尔掩码，标记通过的样本
mask = data.loc[:,"pass"]==1
print(mask)
```

```python
#visualize the data # 可视化通过和未通过的样本
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure()
passed = plt.scatter(data.loc[:, "test1"][mask], data.loc[:, "test2"][mask])
failed = plt.scatter(data.loc[:, "test1"][~mask], data.loc[:, "test2"][~mask])
plt.title("test1-test2")
plt.xlabel("test1")
plt.ylabel("test2")
plt.legend((passed, failed),("passed", "failed"))
plt.show()
```

```python
#define X , y   # 准备特征和标签
x = data.drop(["pass"] , axis = 1)
y = data.loc[: , "pass"]
x1 = data.loc[:,"test1"]
x2 = data.loc[:,"test2"]
```

```python
#create new data   # 创建二阶特征
x1_2 = x1*x1
x2_2 = x2*x2
x1_x2 = x1*x2
x_new = {"x1":x1 , "x2":x2 , "x1_2":x1_2 , "x2_2":x2_2 , "x1_x2":x1_x2}
x_new = pd.DataFrame(x_new)
print(x_new)
```

```python
#establish the model and train it  # 训练逻辑回归模型
from sklearn.linear_model import LogisticRegression
lr2 = LogisticRegression()
lr2.fit(x_new,y)
```

```python
#show the predictive result and its accuracy # 评估模型性能
from sklearn.metrics import accuracy_score
y2_predict = lr2.predict(x_new)
accuracy2 = accuracy_score(y , y2_predict)
print(accuracy2)
```

```python
# 获取逻辑回归模型的参数
x1_new = x1.sort_values()
theta0 = lr2.intercept_
theta1,theta2,theta3,theta4,theta5=lr2.coef_[0][0] , lr2.coef_[0][1] ,lr2.coef_[0][2] , lr2.coef_[0][3],lr2.coef_[0][4]



# 计算决策边界
a = theta4
b = (theta5*x1_new+theta2)
c = theta0 + theta1*x1_new + theta3*x1_new * x1_new
x2_new_boundary = (-b + np.sqrt(b*b - 4*a*c))/(2*a)

print(theta0 , theta1,theta2,theta3,theta4,theta5)


# 可视化通过和未通过的样本及决策边界
fig2 = plt.figure()
passed = plt.scatter(data.loc[:, "test1"][mask], data.loc[:, "test2"][mask])
failed = plt.scatter(data.loc[:, "test1"][~mask], data.loc[:, "test2"][~mask])
plt.plot(x1_new , x2_new_boundary)
plt.title("test1-test2")
plt.xlabel("test1")
plt.ylabel("test2")
plt.legend((passed, failed),("passed", "failed"))
plt.show()
```

### 2.以函数的方式求解边界曲线；

```python
# define f(x) # 定义求解二阶边界的函数
def f(x):
    a = theta4
    b = (theta5*x+theta2)
    c = theta0 + theta1*x + theta3*x * x
    x2_new_boundary1 = (-b + np.sqrt(b*b - 4*a*c))/(2*a)
    x2_new_boundary2 = (-b - np.sqrt(b*b - 4*a*c))/(2*a)
    return x2_new_boundary1,x2_new_boundary2
    
```

```python
# 计算完整的决策边界曲线
x2_new_boundary1 = []
x2_new_boundary2 = []
for x in x1_new:
    x2_new_boundary1.append(f(x)[0])
    x2_new_boundary2.append(f(x)[1])
print(x2_new_boundary1,x2_new_boundary2)
    
```

```python
# 可视化完整的决策边界曲线
fig3 = plt.figure()
passed = plt.scatter(data.loc[:, "test1"][mask], data.loc[:, "test2"][mask])
failed = plt.scatter(data.loc[:, "test1"][~mask], data.loc[:, "test2"][~mask])
plt.plot(x1_new , x2_new_boundary1)
plt.plot(x1_new , x2_new_boundary2)
plt.title("test1-test2")
plt.xlabel("test1")
plt.ylabel("test2")
plt.legend((passed, failed),("passed", "failed"))
plt.show()
```

#### 3.描绘出完整的决策边界曲线

```python
# 生成新的x1范围用于绘制更精细的决策边界
x1_range = [-0.9 + x/10000 for x in range(0,20000)]
x1_range = np.array(x1_range)
x2_new_boundary1 = []
x2_new_boundary2 = []
for x in x1_range:
    x2_new_boundary1.append(f(x)[0])
    x2_new_boundary2.append(f(x)[1])
```

```python
# 可视化完整的最终精细的决策边界曲线
fig4 = plt.figure()
passed = plt.scatter(data.loc[:, "test1"][mask], data.loc[:, "test2"][mask])
failed = plt.scatter(data.loc[:, "test1"][~mask], data.loc[:, "test2"][~mask])
plt.plot(x1_range , x2_new_boundary1)
plt.plot(x1_range , x2_new_boundary2)
plt.title("test1-test2")
plt.xlabel("test1")
plt.ylabel("test2")
plt.legend((passed, failed),("passed", "failed"))
plt.show()
```
